```python
import numpy as np 
import scipy.stats as st 
import matplotlib.pyplot as plt 
```

```python
xs = np.linspace(-5,10,200)
ks = np.arange(50)
```

```python
ks
```

# doing the discrete probablity mass functions. 
### Binomial distribution 
### Poission distributions


```python
pmf_binomial = st.binom.pmf(ks,50,0.25)
# in this case the parameters took by the binomial distributions are
# ks -- > values 
# we have 50 events and each event has 0.25 chance of success

plt.bar(ks, pmf_binomial, label = "Binomial Example", alpha = 0.8)
```

```python
pmf_poission = st.poisson.pmf(ks,10)
plt.bar(ks, pmf_poission, label = "Poission Example", alpha = 0.8)
```
